## Zadanie 1. Licznik laboratoryjny - pierwsza własna klasa

```python
class Counter:
    def __init__(self):
        self.value = 0

    def up(self):
        self.value += 1

    def down(self):
        self.value -= 1

    def zero(self):
        self.value = 0
    
    def get_value(self):
        print(self.value)
        return self.value
```

```python
x = Counter()

x.get_value()
x.up()
x.get_value()
x.up()
x.get_value()
x.down()
x.get_value()
x.zero()
x.get_value()
```

```python
from dataclasses import dataclass
```

```python
m1 = Measurement(1,5.6,'V')
m2 = Measurement(2,6.7,'V')
m3 = Measurement(3,7.8,'V')
m4 = Measurement(4,0.1,'V')

data = [m1, m2, m3, m4]

for i in data:
    print(i)
```

## Zadanie 3. Wektor 2D - metody specjalne i czytelny zapis obiektu

```python
class Vectot2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f'[{self.x},{self.y}]'
    
    def __add__(self, other_vector):
        return Vectot2D(self.x + other_vector.x, self.y + other_vector.y)
    
    def __sub__(self, other_vector):
        return Vectot2D(self.x - other_vector.x, self.y - other_vector.y)
    
    def __mul__(self, thing):
        if (type(thing) == int) or (type(thing) == float):
            return Vectot2D(self.x * thing, self.y * thing)
        if type(thing) == Vectot2D: # iloczyn skalarny wektorów
            return self.x * thing.x + self.y * thing.y
    
    def __rmul__(self, f):
        return Vectot2D(self.x * f, self.y * f)
    
    def lenght(self):
        return (self.x**2 + self.y**2)**0.5
```

```python
v1 = Vectot2D(1,2)
v2 = Vectot2D(-6,-9)
print(v1,v2)
```

```python
print(v1+v2)
print(v1-v2)
print(v1*3)
print(2*v2)
print(v1*v2)
print(v2.lenght())
```

## Zadanie 4. Pojemnik na wyniki pomiarów - indeksowanie i iteracja

```python
class MeasurementSeries:
    def __init__(self, values):
        self.values = values

    def __getitem__(self, key):
        return self.values[key]
    
    def __setitem__(self, key, new_value):
        self.values[key] = new_value

    def __len__(self):
        return len(self.values)
    
    def __iter__(self):
        for i in self.values:
            yield i
    
    def avg(self):
        return sum(self.values) / len(self.values)
    
    def max(self):
        max_val = self.values[0]
        for i in self.values:
            if max_val < i:
                max_val = i
        return max_val
    
    def min(self):
        min_val = self.values[0]
        for i in self.values:
            if min_val > i:
                min_val = i
        return min_val
        
```

```python
data1 = [1,2,3,4,11,2,3,4]
ms1 = MeasurementSeries(data1)
print(ms1[0])
ms1[0] = 99
print(ms1[0])
print(len(ms1))
```

```python
for i in ms1:
    print(i)
```

```python
print(ms1.avg(), ms1.max(), ms1.min())
```

## Zadanie 5. Funkcja wielomianowa - obiekt funkcyjny

```python
class Polynomial:
    def __init__(self, coef):
        self.coef = coef

    def __call__(self, val):
        c = Counter()
        suma = 0
        for i in self.coef:
            suma += i * (val**c.get_value())
            c.up()
        return suma
    
    def dirivative(self, val):
        c = Counter()
        suma = 0
        for  i in self.coef:
            suma += i * c.get_value() * val**(c.get_value() - 1)
            c.up()
        return suma
```

```python
f = Polynomial([1,1,1])
print(f(-3))
print(f.dirivative(-3))
```

## Zadanie 6. Konto punktowe gracza - dziedziczenie i `super()`

```python
class Account:
    def __init__(self, name = 'placeholder_name', points = 0):
        self.name = name
        self.points = points

    def add_points(self, val):
        self.points += val

    def sub_points(self, val):
        self.points -= val

    def __str__(self):
        return f'Nazwa konta: {self.name}, punkty: {self.points}'
    
class PremiumAccount(Account):
    def __init__(self, name='placeholder_name', points=0, point_bonus=2):
        super().__init__(name, points)
        self.point_bonus = point_bonus

    def add_points(self, val):
        super().add_points(val * self.point_bonus)

    def __str__(self):
        return f'{super().__str__()}, bonus: {self.point_bonus}'
    
class PenaltyAccount(Account):
    def __init__(self, name='placeholder_name', points=0, penalty=1):
        super().__init__(name, points)
        self.penalty = penalty

    def add_points(self, val):
        print(f'Pobrano {self.penalty} puntk(y/ów) za wykonanie tej operacji (dodanie punktów)')
        super().add_points(val - self.penalty)

    def sub_points(self, val):
        print(f'Pobrano {self.penalty} puntk(y/ów) za wykonanie tej operacji (odjęcie punktów)')
        super().sub_points(val + self.penalty)

    def __str__(self):
        print(f'Pobrano {self.penalty} puntk(y/ów) za wykonanie tej operacji (wypisanie stanu konta)')
        super().sub_points(self.penalty)
        return f'{super().__str__()}, opłata karna: {self.penalty}'
        
```

```python
acc1 = PremiumAccount('marek')
print(acc1)
acc1.add_points(10)
print(acc1)
acc1.sub_points(1)
print(acc1)
```

```python
acc2 = PenaltyAccount('janusz',10)
print(acc2)
acc2.add_points(10)
acc2.sub_points(2)
print(acc2)
```

## Zadanie 7. Zwierzęta na planszy - polimorfizm

```python
class Animal:
    def __init__(self, name, x=0, y=0):
        self.name = name
        self.x = x
        self.y = y

    def move(self, new_x, new_y):
        self.x = new_x
        self.y = new_y

    def sound(self):
        print('*default animal sound*')

    def info(self):
        print(f'name = {self.name}, x = {self.x}, y = {self.y}')
```

```python
import random
class Dog(Animal):
    def __init__(self, name, x=0, y=0):
        super().__init__(name, x, y)

    def sound(self):
        r = random.randint(1,3)
        if r == 1:
            print('*woof*')
        elif r == 2:
            print('*bark*')
        else:
            print('*grrr*')

    def move(self, suggest_x, suggest_y):
        rx = random.randint(-1,1)
        ry = random.randint(-1,1)
        self.x += rx
        self.y += ry
```

```python
import random
class Bird(Animal):
    def __init__(self, name, x=0, y=0):
        super().__init__(name, x, y)

    def sound(self):
        r1 = random.randint(0,2)
        r2 = random.randint(4,7)
        s = ['c','a','w',' ','c','a','a','w']

        for i in range(len(s)):
            if (i == r1) or (i == r2):
                s[i] = s[i].upper()
        
        print(''.join(s))

    def move(self, add_x, add_y):
        self.x += add_x
        self.y += add_y
        print('*wooosh*')
```

```python
a = Animal('john mammal',1,3)
b = Bird('birb',3,7)
d = Dog('dawg',-1, 10)

animals = [a,b,d]

for i in animals:
    i.info()
    i.sound()
    i.move(2,-4)
    i.info()
    print('\n')
```
